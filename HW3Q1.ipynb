```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# "the default sans-serif font is Arial"
matplotlib.rcParams['font.sans-serif'] = "Arial"
# Then, "ALWAYS use sans-serif fonts"
matplotlib.rcParams['font.family'] = "sans-serif"
matplotlib.rcParams.update({'font.size': 10})
```

```python
# Define a permitation function to calculate x!
def permu(x):
    function = 1
    for i in range(1, x+1):
        function *= i
    return function
```

```python
# Define the MRL function 
def MRLfunction(t, k, kc, n):
    # Calculate the integral part
    function = 0
    for i in range(n+1):
        function += ((1/kc) * (-1)**i * (kc*t)**(2*i+1) / ((2*i+1)*permu(i)))
    # Calculate MRL
    mrl = np.exp(k*t**2/2) * (np.sqrt(np.pi)/(kc*2) - function)
    return mrl 
```

```python
K = 2e-6
KC = 1e-3
N = 160
```

```python
# Plot MRL(t) for different t
T = np.linspace(0, 5000, 101)
MRL = np.zeros(len(T))
for j in range(len(T)):
    MRL[j] = MRLfunction(T[j], K, KC, N)
```

```python
plt.plot(T, MRL, color='maroon', linewidth = 1, marker = 'o', markersize = 3)
plt.ylim(0, 900)
plt.xlim(0, 4000)
plt.xlabel('$t$')
plt.ylabel('$MRL$')
plt.rcParams['figure.figsize'] = (6, 4) 
plt.rcParams['savefig.dpi'] = 300 
plt.rcParams['figure.dpi'] = 300 
plt.savefig(r"%s.jpg"%('MRL'), bbox_inches='tight')
plt.show()
```
