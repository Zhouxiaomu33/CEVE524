```python
from scipy import stats
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# "the default sans-serif font is Arial"
matplotlib.rcParams['font.sans-serif'] = "Arial"
# Then, "ALWAYS use sans-serif fonts"
matplotlib.rcParams['font.family'] = "sans-serif"
matplotlib.rcParams.update({'font.size': 10})
```

```python
# Prepare the original data
Data = pd.read_csv('OriginalHurricanData.csv')
```

```python
Hval = np.array(Data['Number'])
HPMF = np.array(Data['Frequency']/sum(Data['Frequency']))
```

```python
# Plot the PMF
PMF = stats.rv_discrete(values=(Hval, HPMF))
fig, ax = plt.subplots(1, 1)
ax.plot(Hval, PMF.pmf(Hval), 'o', ms = 5, mec = 'maroon')
ax.vlines(Hval, 0, PMF.pmf(Hval), colors = 'maroon', linestyles='-', lw=2)
ax.set_xlim( (0, 32) )
ax.set_ylim( (0, 0.12) )
# plt.title('PMF of $H$')
plt.ylabel('Probability')
plt.rcParams['figure.figsize'] = (6, 4) 
plt.rcParams['savefig.dpi'] = 300 
plt.rcParams['figure.dpi'] = 300 
plt.savefig(r"%s.jpg"%('PMF'), bbox_inches='tight')
plt.show()
```

```python
# b calculate E(H)
EH = np.sum(Hval*HPMF)
EH
```

```python
# c calculate V(H)
EHS = np.sum((Hval*Hval)*HPMF)
VH = EHS - (EH*EH)
VH
```
